```python
import xml.etree.ElementTree as ET


root = ET.Element("students")


sv1 = ET.SubElement(root, "student", id="SV001")
ET.SubElement(sv1, "name").text = "Nguyen Van A"
ET.SubElement(sv1, "year").text = "2002"
ET.SubElement(sv1, "class").text = "CNTT1"
ET.SubElement(sv1, "gender").text = "Nam"


sv2 = ET.SubElement(root, "student", id="SV002")
ET.SubElement(sv2, "name").text = "Tran Thi B"
ET.SubElement(sv2, "year").text = "2003"
ET.SubElement(sv2, "class").text = "CNTT2"
ET.SubElement(sv2, "gender").text = "Nữ"


tree = ET.ElementTree(root)
tree.write("students.xml", encoding="utf-8", xml_declaration=True)

print(" Đã tạo file students.xml thành công!")
```

```python
from xml.dom import minidom


doc = minidom.parse("sample.xml")


root = doc.documentElement
print("Tên thẻ gốc:", root.tagName)


company_name = root.getElementsByTagName("name")[0].firstChild.data
print("Tên công ty:", company_name)


staffs = root.getElementsByTagName("staff")
print("\nDanh sách nhân viên:")
for staff in staffs:
    staff_id = staff.getAttribute("id")
    name = staff.getElementsByTagName("name")[0].firstChild.data
    salary = staff.getElementsByTagName("salary")[0].firstChild.data
    print(f"- ID: {staff_id}, Tên: {name}, Lương: {salary}")
```

```python
import requests
import csv
import xml.etree.ElementTree as ET


url = "https://vnexpress.net/rss/tin-moi-nhat.rss"


response = requests.get(url)
response.encoding = 'utf-8' 


root = ET.fromstring(response.text)


with open("rss_data.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "Link"])  


    for item in root.findall(".//item"):
        title = item.find("title").text if item.find("title") is not None else ""
        link = item.find("link").text if item.find("link") is not None else ""

  
        writer.writerow([title, link])

print(" Đã ghi dữ liệu RSS vào file rss_data.csv thành công!")
```

```python
import json

def thong_ke_nhan_vien(filename):

    with open(filename, 'r', encoding='utf-8') as f:
        data = json.load(f)

 
    ten_cty = data['ten_cong_ty']
    dia_chi = data['dia_chi']
    don_vi_list = data['don_vi']


    tong_nv = sum(dv['so_nhan_vien'] for dv in don_vi_list)


    print(f"Tên công ty: {ten_cty}")
    print(f"Địa chỉ: {dia_chi}")
    print("----- Thống kê nhân viên theo đơn vị -----")


    for i, dv in enumerate(don_vi_list, start=1):
        ty_le = dv['so_nhan_vien'] / tong_nv * 100
        print(f"{i}. Tên đơn vị: {dv['ten_don_vi']}")
        print(f"   - Số nhân viên: {dv['so_nhan_vien']}")
        print(f"   - Tỷ lệ: {ty_le:.2f}%")
        print()

if __name__ == "__main__":
    thong_ke_nhan_vien("nhanvien.json")
```

```python
import json
from datetime import datetime

def nhap_giao_dich():
    ds_giao_dich = []
    print("=== NHẬP GIAO DỊCH TIỀN & VÀNG ===")
    while True:
        loai = input("Nhập loại giao dịch (tiền/vàng): ").strip()
        so_tien = float(input("Nhập số tiền: "))
        mo_ta = input("Nhập mô tả: ")

        giao_dich = {
            "loai": loai,
            "so_tien": so_tien,
            "mo_ta": mo_ta
        }
        ds_giao_dich.append(giao_dich)

        tiep_tuc = input("Nhập thêm giao dịch? (y/n): ").strip().lower()
        if tiep_tuc != 'y':
            break
    return ds_giao_dich


def ghi_file_json(ds_giao_dich):

    now = datetime.now()
    ten_file = now.strftime("%Y-%m-%d-%H-%M-%S.json")


    with open(ten_file, "w", encoding="utf-8") as f:
        json.dump(ds_giao_dich, f, ensure_ascii=False, indent=4)

    print(f"Đã ghi {len(ds_giao_dich)} giao dịch vào file '{ten_file}'.")


def main():
    ds = nhap_giao_dich()
    chon = input("Bạn có muốn ghi vào file JSON không? (1: Có / 0: Không): ").strip()
    if chon == "1":
        ghi_file_json(ds)
    else:
        print(" Không ghi dữ liệu.")


if __name__ == "__main__":
    main()
```

```python
import requests

def doc_post_tu_api():
    url = "https://jsonplaceholder.typicode.com/posts"

  
    print(" Đang đọc dữ liệu từ API...")
    response = requests.get(url)


    if response.status_code != 200:
        print(" Lỗi khi truy cập API!")
        return


    ds_posts = response.json()

 
    print(f"\nTổng số bài post: {len(ds_posts)}\n")


    for post in ds_posts:
        print(f"UserID: {post['userId']}")
        print(f"ID: {post['id']}")
        print(f"Title: {post['title']}")
        print(f"Body: {post['body']}")
        print("-" * 50)

if __name__ == "__main__":
    doc_post_tu_api()
```

```python
import requests

def doc_sach_noi_bat():
    url = "https://jsonplaceholder.typicode.com/comments?postId=1"
    print(" Đang đọc dữ liệu sách nổi bật từ API...")

    response = requests.get(url)

    if response.status_code != 200:
        print(" Lỗi khi truy cập API!")
        return

    ds_bai_post = response.json()

    print(f"\nTổng số bài post nổi bật: {len(ds_bai_post)}")
    print("=== Danh sách 3 bài post nổi bật đầu tiên ===")

    for post in ds_bai_post[:3]:
        print(f"\nID: {post['id']}")
        print(f"Tên người đăng: {post['name']}")
        print(f"Email: {post['email']}")
        print(f"Nội dung: {post['body']}")
        print("-" * 60)

if __name__ == "__main__":
    doc_sach_noi_bat()
```
